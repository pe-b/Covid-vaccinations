```python
import findspark
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import desc
from pyspark.sql.functions import asc

findspark.init()
```

```python
# Configure the spark session
spark = SparkSession \
    .builder \
    .appName("Covid Vaccines") \
    .getOrCreate()
```

```python
# Load the data
df = spark.read.option("header", True).option("inferSchema", True).csv("country_vaccinations.csv")
```

```python
# Check the schema
df.printSchema()
```

```python
# Check data date range
min_date = df.agg({"date" : "min"}).collect()[0][0]
max_date = df.agg({"date" : "max"}).collect()[0][0]

print("First day in the records: " + min_date)
print("Last day in the records: " + max_date)
```

```python
# Total vaccinations by country high to low
df.groupBy("country").sum("daily_vaccinations").sort(desc("sum(daily_vaccinations)")).withColumnRenamed("sum(daily_vaccinations)", "total_vaccinations").show()
```

```python
# Total vaccinations by country high to low
df.groupBy("country").sum("daily_vaccinations").sort(asc("sum(daily_vaccinations)")).withColumnRenamed("sum(daily_vaccinations)", "total_vaccinations").show()
```
